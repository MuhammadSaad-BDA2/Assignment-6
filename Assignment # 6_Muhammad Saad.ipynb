```
Emp_Data = spark\
            .read\
            .format('csv')\
            .option("inferSchema","true")\
            .option("header","true")\
            .load("/FileStore/tables/employee_earnings_report_2016_1.csv")
```

```
data=Emp_Data.crosstab('TITLE','Gender')
data.show()
data.write.format("csv").save("/FileStore/tables/dataframe_wa.csv")
```

```
postal_count=Emp_Data.groupby('POSTAL','DEPARTMENT_NAME',"TITLE").count().sort("count", ascending=False)
postal_count.show(3,truncate=False)

postal_count.write.mode('append').format("csv").save("/FileStore/tables/dataframe_wa.csv")
```

```
Emp_Data.createOrReplaceTempView("Emp_Data")
```

```
import pyspark.sql.functions as spark_func
import pyspark.sql.types as spark_types
import pyspark.sql.functions as F

Emp_Data.groupBy('TITLE','DEPARTMENT_NAME','Gender').agg(F.count('Gender').alias('Male')).where(F.col('Gender')== 'M').show()
Emp_Data.groupBy('TITLE','DEPARTMENT_NAME','Gender').agg(F.count('Gender').alias('Female')).where(F.col('Gender')== 'F').show()

```

```
Emp_Data.groupBy('TITLE','DEPARTMENT_NAME','POSTAL').agg(F.count('POSTAL').alias('Count')).orderBy(F.col('Count').desc()).show(3)
```
